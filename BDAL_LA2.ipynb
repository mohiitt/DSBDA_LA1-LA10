```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('academic.csv')
df
```

```python
df.isnull().sum()
```

```python
missing_values_format=['NaN','na','N.A',]
df.replace(missing_values_format,np.nan,inplace=True)
df
```

```python
#type 1 to handle missing values
df2=df.dropna()
df2
```

```python
#type 2
df['SPOS'].fillna(method='pad',inplace=True)
df
#pad replaces missing values from value of previous row
```

```python
df['WT'].fillna(method='bfill',inplace=True)
df
#bfill replacess missing values from value of next row
```

```python
df
```

```python
sns.boxplot(data=df,y='SPOS')
```

```python
sns.boxplot(data=df,x='WT')
```

```python
df['WT']=np.where((df['WT']>100),(df['WT'].median()),(df['WT']))
sns.boxplot(data=df,x='WT')
```

```python
from sklearn.preprocessing import StandardScaler

# Assuming your DataFrame is named 'df' and your variable is 'DSBDA'
scaler = StandardScaler()
df['DSBDA_scaled'] = scaler.fit_transform(df[['DSBDA']])  # Fit and transform for standardization
df['DSBDA_scaled']
```

```python
from sklearn.preprocessing import MinMaxScaler

# Assuming your DataFrame is named 'df' and your variable is 'DSBDA'
scaler = MinMaxScaler(feature_range=(0, 1))  # Specify range (default 0-1)
df['WT_scaled'] = scaler.fit_transform(df[['WT']])  # Fit and transform for scaling

df['WT_scaled']
```
